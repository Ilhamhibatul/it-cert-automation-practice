```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import tree
```

# Dataset Wine

```python
data_wine = load_wine()
wine = pd.DataFrame(data_wine.data)
print("Features Name : ", data_wine.feature_names)
print("Dataset Shape : ", wine.shape)
print("Dataset : \n", wine.head())
```

Penjelasan : pada dataset wine ini kita terlebih dulu loadwine yang ditampung dalam variabel data_wine. Kemudian variabel sebelumnya kita beri .data dan kita tampung dalam variabel wine. untuk menampilkan features namenya kita bisa beri perintah .feature_name, untuk menampilkan shapenya, kita bisa menggunakan .shape,dan untuk menampilkan datasetnya kita bisa gunakan .head. Jika dijalankan, akan menampilkan feature name dari dataset wine tersebut, kemudian untuk shapenya ada 17 baris dan 13 kolom   

```python
X = wine.values[:, 0:13]
Y = data_wine.target

# Untuk split data
x_train1, x_test1, y_train1, y_test1 = train_test_split(X, Y, test_size= 0.2, random_state= 100)
x_train2, x_test2, y_train2, y_test2 = train_test_split(X, Y, test_size= 0.3, random_state= 100)
x_train3, x_test3, y_train3, y_test3 = train_test_split(X, Y, test_size= 0.4, random_state= 100)
```

Selanjutnya, kita definisikan nilai X dan Y. Nilai X kita dapat dari wine.values dengan index slicing dari kolom index ke 0 sampai 13. Untuk Y kita menggunakan data_wine.target untuk menampung targetnya. Lalu untuk splitnya kita bisa memakai train_test_split dan disesuaikan dengan keinginan, test size kita sesuaikan dengan presentase 20%, 30%, dan 40%.


```python
clf = DecisionTreeClassifier(random_state=100)
clf.fit(x_train1, y_train1)

y_pred1 = clf.predict(x_test1)
print("Data testing 20% : \n", y_pred1)

cm1 = confusion_matrix(y_test1, y_pred1);
print("\nConfusion matrix data testing 20% : \n ", cm1)

accuracy1= accuracy_score(y_test1, y_pred1);
print("\nScore akurasi data testing 20% : \n ",accuracy1)

tree.plot_tree(clf)
text_representation = tree.export_text(clf)
print("\nTree : \n", text_representation)
```

Penjelasan : Selanjutnya kita definisikan decision trree classificationnya dengan random_statenya = 100 dan ditampung dalam variabel clf. Lalu untuk mendapatkan prediksinya kita bisa menggunakan clf.predict dengan valuesnya dari x_test1. Lalu untuk mendapatkan confusion matrixnya kita bisa menggunakan values y_test1, dan y_pred1. Lalu utuk mendapat skor akurasinya kita menggunakan value y_test1 dan y_pred1. Lalu untuk menampilkan treenya kita bisa memakai tree.plot_tree dan masukkan variabel clfnya.

```python
clf = DecisionTreeClassifier(random_state=100)
clf.fit(x_train2, y_train2)

y_pred2 = clf.predict(x_test2)
print("Data testing 30% : \n", y_pred2)

cm2 = confusion_matrix(y_test2, y_pred2);
print("\nConfusion matrix data testing 30% : \n ", cm2)

accuracy2= accuracy_score(y_test2, y_pred2);
print("\nScore akurasi data testing 30% : \n ",accuracy2)

tree.plot_tree(clf)
text_representation = tree.export_text(clf)
print("\nTree : \n", text_representation)
```

Penjelasan : lalu untuk data testing 30%, pengerjaannya sama seperti 20% lalu yang membedakan hanya penggunaan variabel yang menampung data testing 30%nya.

```python
clf = DecisionTreeClassifier(random_state=100)
clf.fit(x_train3, y_train3)

y_pred3 = clf.predict(x_test3)
print("Data testing 40% : \n", y_pred3)

cm3 = confusion_matrix(y_test3, y_pred3);
print("\nConfusion matrix data testing 40% : \n ", cm3)

accuracy3= accuracy_score(y_test3, y_pred3);
print("\nScore akurasi data testing 40% : \n ",accuracy3)

tree.plot_tree(clf)
text_representation = tree.export_text(clf)
print("\nTree : \n", text_representation)
```

# Dataset Pima Indian Diabetes

```python
diabetes = pd.read_csv("pima-indians-diabetes.csv")
print("Features Name : ", diabetes.columns)
print("Dataset Shape : ", diabetes.shape)
print("Dataset : \n", diabetes.head())
```

pada dataset pima indian diabetes ini kita terlebih dulu read file datasetnya dan ditampung dalam variabel diabetes. Untuk menampilkan features namenya kita bisa beri perintah .columns, untuk menampilkan shapenya, kita bisa menggunakan .shape,dan untuk menampilkan datasetnya kita bisa gunakan .head. Lalu jika dijalankan akan menampilkan feature name, kemudian shapenya memiliki 768 baris dan 9 kolom, dan menampilkan juga datasetnya.

Penjelasan : lalu untuk penggunaan variabel X dan Y didapat dari diabetes.iloc dengan menggunakan index slicing kemudian kita split datanya sama seperti sebelumnya.

```python
clf = DecisionTreeClassifier(random_state=100)
clf.fit(x_train1, y_train1)

y_pred1 = clf.predict(x_test1)
print("Data testing 20% : \n", y_pred1)

cm1 = confusion_matrix(y_test1, y_pred1);
print("\nConfusion matrix data testing 20% : \n ", cm1)

accuracy1= accuracy_score(y_test1, y_pred1);
print("\nScore akurasi data testing 20% : \n ",accuracy1)

tree.plot_tree(clf)
text_representation = tree.export_text(clf)
print("\nTree : \n", text_representation)
```
